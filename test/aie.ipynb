```python
import aie

aie.Authenticate()
aie.Initialize()
```

```python
image = aie.Image('LC08_L2SP_119039_20180311_20200901_02_T1')

# NDVI运算
result_image = image.normalizedDifference(['SR_B5', 'SR_B4'])

map = aie.Map(
    center=result_image.getCenter(),
    height=800,
    zoom=8
)

vis_params = {
    'min': -1,
    'max': 1,
    'palette': [ 
        '#FFFFFF', '#CE7E45', '#DF923D', '#F1B555', '#FCD163', '#99B718',
        '#74A901', '#66A000', '#529400', '#3E8601', '#207401', '#056201',
        '#004C00', '#023B01', '#012E01', '#011D01', '#011301'
    ]
}

map.addLayer(
    result_image,
    vis_params,
    'ndvi result image',
    bounds=result_image.getBounds()
)
map
```

```python
from aie.map.aie_layer import AIEImageLayer

aie_il = AIEImageLayer('ndvi result image', result_image, vis_params, bounds=result_image.getBounds())
```

```python
from folium import TileLayer
import xyzservices

basemap = aie_il

if isinstance(basemap, dict):
    layer = leafmap.basemap_to_tiles(basemap)
    print(layer)
            
if isinstance(basemap, xyzservices.lib.TileProvider):
    url = basemap.build_url()
elif isinstance(basemap, dict):
    url = basemap.get("url", "")
else:
    raise ValueError("Invalid basemap type")

tl = TileLayer(
    url=url,
    max_zoom=basemap.get('max_zoom', 19),
    min_zoom=basemap.get('min_zoom', 1),
    attribution=basemap.get('html_attribution', '') or basemap.get('attribution', ''),
    name=basemap.get('name', '')
)
```
